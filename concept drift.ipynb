```python
st.norm.ppf(0.01)
```

```python
np.random.seed(42)

fig, ax = plt.subplots(1, 2, figsize=(12,4), dpi=100)

x = np.linspace(st.norm.ppf(0.01), st.norm.ppf(0.99), 1000)
ax[0].plot(x, st.norm.pdf(x), 'r-', lw=5, alpha=0.6, label='Class 0')
ax[0].plot((x+2)*2, st.norm.pdf(x)/2, 'o-', lw=5, alpha=0.6, label='Class 1')
ax[0].legend()
ax[0].set_title('Initial distribution')
ax[0].axvline(1.7, color='r')

ax[1].plot((x)*2, st.skewnorm.pdf(x, 3), 'r-', lw=5, alpha=0.6, label='Class 0')
ax[1].plot((x)*2, st.skewnorm.pdf(x, -2)/2, 'o-', lw=5, alpha=0.6, label='Class 1')
ax[1].legend()
ax[1].set_title('Posthoc distribution')
ax[1].axvline(1.7, color='r')

plt.legend()
plt.show()
```
